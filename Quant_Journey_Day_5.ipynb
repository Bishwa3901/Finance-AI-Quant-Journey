```python
import yfinance as yf
import pandas as pd

stock = yf.download("RELIANCE.NS", start="2015-01-01", end="2024-01-01")

print(stock.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(stock['Close'])
plt.title("Reliance Stock Price")
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```

```python
stock['Return'] = stock['Close'].pct_change()

stock.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(stock['Return'])
plt.title("Daily Returns")
plt.show()
```

Measuring Volatility

```python
stock['Return'].std()
```

```python
#Measuring Rolling Volatility
stock['Volatility'] = stock['Return'].rolling(30).std()

plt.figure(figsize=(10,5))
plt.plot(stock['Volatility'])
plt.title("30-Day Rolling Volatility")
plt.show()
```

Moving Averages

```python
stock['MA50'] = stock['Close'].rolling(50).mean()
stock['MA200'] = stock['Close'].rolling(200).mean()
```

```python
plt.figure(figsize=(12,6))

plt.plot(stock['Close'], label='Price')
plt.plot(stock['MA50'], label='MA50')
plt.plot(stock['MA200'], label='MA200')

plt.legend()
plt.show()
```
